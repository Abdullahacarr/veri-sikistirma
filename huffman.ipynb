```python
from heapq import heappush, heappop, heapify
from collections import defaultdict
import timeit
def encode(symb2freq):
    heap = [[wt, [sym, ""]] for sym, wt in symb2freq.items()]
    heapify(heap)
    while len(heap) > 1:
        lo = heappop(heap)
        hi = heappop(heap)
        for pair in lo[1:]:
            pair[1] = '0' + pair[1]
        for pair in hi[1:]:
            pair[1] = '1' + pair[1]
            
        heappush(heap, [lo[0] + hi[0]] + lo[1:] + hi[1:])
    return sorted(heappop(heap)[1:], key=lambda p: (len(p[-1]), p))
ab=input("Metni giriniz :") 
symb2freq = defaultdict(int)
liste=list(ab)
dict = dict.fromkeys(liste)
j=0
k=2
for ka in ab:
    symb2freq[ka] += 1
    d=ab[j:k]
    if(ab.count(d)>30):
        dict[d] = ab.count(d)
    j+=1
    k+=1
huff = encode(symb2freq)
value=[]
key=[]
for k in huff:
    value.append(k[1])
    key.append(k[0])
res = {key[i]: value[i] for i in range(len(key))}
k=""
g=0
f=2
uzz=len(liste)
t=0
while (t<uzz):
    r=ab[g:f]
    a=ab[t]
    if r in res:
        k +=res[r]
        t+=2
        g+=2
        f+=2
    else:
        k +=res[a]
        t+=1
        g+=1
        f+=1
    
print(k)
print("sktrlan bit uzunluu :" ,len(k))
print("orjinal metin bit says :",len(ab)*8)
timeit.timeit('"-".join(str(n) for n in range(100))', number=10000)
```
